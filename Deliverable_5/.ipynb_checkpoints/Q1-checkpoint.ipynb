```python
%matplotlib inline
import cv2  
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
# read image
# image_path = os.path.join('..', 'Goldhill_2-1.jpg')
goldhill_orig_path = 'Goldhill_2-1.jpg'
goldhill_orig_img = cv2.imread (goldhill_orig_path)

goldhill_noise_path = 'Goldhill_2noisy.jpg'
goldhill_noise_img = cv2.imread(goldhill_noise_path)

plt.imshow(goldhill_orig_img)
plt.show()


# cv2.imshow("Original Image", goldhill_orig_img)
# cv2.imshow("Noisy Image", goldhill_noise_img)


# cv2.waitKey(0)
# cv2.destroyAllWindows()
# write image
# cv2.imwrite(os.path.join('..', 'data', 'sample1_copy.jpg'), img)

# visualize images
```

```python
#show noisy image
plt.imshow(goldhill_noise_img)
plt.show()
```

```python
#Apply Average Technique
k_size = 5
blurred_img = cv2.blur(goldhill_noise_img, (k_size, k_size))
plt.imshow(blurred_img)
plt.show()

```

```python
#Apply Guassian Technique
GaussianBlur_img = cv2.GaussianBlur(goldhill_noise_img, (k_size, k_size), 0)
plt.imshow(GaussianBlur_img)
plt.show()
```

```python
#Apply median Technique
medianBlur_img = cv2.medianBlur(goldhill_noise_img, k_size)
plt.imshow(medianBlur_img)
plt.show()
```

```python
# PSNR of Noisy Image vs Original
psnr_org_v_noise = cv2.PSNR(goldhill_orig_img, goldhill_noise_img)
psnr_org_v_ave_blur = cv2.PSNR(goldhill_orig_img, blurred_img)
psnr_org_v_gauss = cv2.PSNR(goldhill_orig_img, GaussianBlur_img)
psnr_org_v_median = cv2.PSNR(goldhill_orig_img, medianBlur_img)


print(psnr_org_v_noise, 'in dB of original vs noisy')
print(psnr_org_v_ave_blur, 'in dB of original vs Averaging Blur')
print(psnr_org_v_gauss, 'in dB of original vs Gaussian Blur')
print(psnr_org_v_median, 'in dB of original vs Median Blur')

```

```python
#Apply Sharpening Technique
# L_ave_img = cv2.Laplacian(blurred_img, cv2.CV_16S)
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
f_ave_img = cv2.filter2D(blurred_img, -1, kernel)
plt.imshow(f_ave_img)
plt.show()
```

```python
#Apply Sharpening Technique
f_gauss_img = cv2.filter2D(GaussianBlur_img, -1, kernel)
plt.imshow(f_gauss_img)
plt.show()
```

```python
#Apply Sharpening Technique
f_median_img = cv2.filter2D(medianBlur_img, -1, kernel)
plt.imshow(f_median_img)
plt.show()
```

```python
psnr_org_v_f_ave_blur = cv2.PSNR(goldhill_orig_img, f_ave_img)
psnr_org_v_f_gauss = cv2.PSNR(goldhill_orig_img, f_gauss_img)
psnr_org_v_f_median = cv2.PSNR(goldhill_orig_img, f_median_img)

print(psnr_org_v_f_ave_blur, 'in dB of original vs Averaging Blur and Sharpening')
print(psnr_org_v_f_gauss, 'in dB of original vs Gaussian Blur and Sharpening')
print(psnr_org_v_f_median, 'in dB of original vs Median Blur and Sharpening')
```
