```python
%matplotlib inline
import cv2  
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
# read image
# image_path = os.path.join('..', 'Goldhill_2-1.jpg')
birds_lake_path = 'birds_lake.jpg'
birds_lake_img = cv2.imread (birds_lake_path)

plt.imshow(birds_lake_img)
plt.show()
```

```python
# crop image
cropped_img = birds_lake_img[25:400, 150:480]
plt.imshow(cropped_img)
plt.show()
```

```python
#convert to grayscale image first
gray_img = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap='gray')
plt.show()

```

```python
# apply thresholding

ret, simple_thresh = cv2.threshold(gray_img, 75, 255, cv2.THRESH_BINARY_INV)

plt.imshow(simple_thresh, cmap='gray')
plt.show()

```

```python
# create contours
contours, hierarchy = cv2.findContours (simple_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

contours_num = 0

for cnt in contours:
    # print (cv2.contourArea(cnt))
    # Removing noise
    if cv2.contourArea(cnt) > 10:
        cv2.drawContours (cropped_img, cnt, -1, (0, 255, 0), 1)
        # Count the counturs!!!
        # Drawing bounding boxes around the objects
        x1, y1, width, height = cv2.boundingRect (cnt)

        cv2.rectangle (cropped_img, (x1, y1), (x1+ width, y1 + height), (0, 255, 0), 2)
        contours_num += 1

print (f"Number of contours: {contours_num}")
# cv2.drawContours(birds_lake_img, contours, -1, (0,255,0), 3)
# plt.imshow(cv2.drawContours(cropped_img, contours, -1, (255,0,0), 3))
plt.imshow(cropped_img)
plt.show()
```

```python
cv2.imwrite('birds_lake_with_contours.jpg', cropped_img)
```
